```python
import google.generativeai as genai
import gradio as gr

# Configure Gemini API
genai.configure(api_key="AQ.Ab8RN6LboVE4ogdNAL4iQDcin96kG9UfPW9-LHTjLtklGxK-Gg")

# Use a supported model
model = genai.GenerativeModel("gemini-2.5-flash")

def saving_assistant(user_input):
    try:
        response = model.generate_content(user_input)

        if hasattr(response, "text") and response.text:
            return response.text
        else:
            return "No response received."

    except Exception as e:
        return f"Error:\n{e}"

with gr.Blocks() as demo:
    gr.Markdown("# 💰 My Saving Assistant")

    user_input = gr.Textbox(
        label="Enter your question",
        placeholder="Example: Give ideas for saving ₹10,000 per month"
    )

    output = gr.Textbox(label="Output", lines=10)

    with gr.Row():
        submit = gr.Button("Submit")
        clear = gr.ClearButton([user_input, output])

    submit.click(
        fn=saving_assistant,
        inputs=user_input,
        outputs=output
    )

demo.launch(debug=True, share=True)
```
